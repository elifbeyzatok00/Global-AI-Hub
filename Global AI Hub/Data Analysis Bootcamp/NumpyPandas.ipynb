```python
x = np.array([[1,2], [3,4]], dtype=np.float64)
y = np.array([[1,2], [3,4]], dtype=np.float64)

print(x + y)
print(np.add(x,y))
```

```python
x = np.array([[1,2], [3,4]], dtype=np.float64)
y = np.array([[1,2], [3,4]], dtype=np.float64)

print(x - y)
print(np.subtract(x,y))
```

```python
print(np.dot(x,y))
```
